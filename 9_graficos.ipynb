Criando gráficos com plotly

```python
import plotly.express as px
```

```python
import pandas as pd

df_vendas = pd.read_csv('vendas.csv', decimal=',', sep=';', index_col=0, parse_dates=True)
df_produtos = pd.read_csv('produtos.csv', decimal=',', sep=';', index_col=0)

df_produtos = df_produtos.rename(columns={'nome': 'produto'})
df_produtos = df_produtos[['produto', 'preco']]
df_vendas = pd.merge(left=df_vendas.reset_index(), 
                     right=df_produtos,
                     on='produto',
                     how='left')

df_vendas = df_vendas.set_index('data')
df_vendas.head(10)
df_vendas
```

Gráficos de linha

```python
df_vendas['dia_venda'] = df_vendas.index.date
df_vendas.head()
```

```python
df_vendas_dia= df_vendas.groupby('dia_venda').agg({'preco': 'sum', 'produto': 'count'})
df_vendas_dia
```

```python
px.line(df_vendas_dia)
```

```python
px.line(df_vendas_dia, y='preco')
```

Criando gráfico de barras

```python
px.bar(df_vendas_dia, y='preco')
```

Gráficos de pizza

```python
px.pie(df_vendas, names='filial', values='preco')
```
