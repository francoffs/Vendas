```python
import pandas as pd

df_vendas = pd.read_csv('vendas.csv', decimal=',', sep=';', index_col=0)

df_produtos = pd.read_csv('produtos.csv', decimal=',', sep=';')
```

Utilizando o método merge

```python
df_produtos = df_produtos.rename(columns={'nome': 'produto'})
df_produtos
```

```python
df_produtos = df_produtos[['produto', 'preco']]
df_produtos
```

```python
df_vendas_mesclado = pd.merge(left=df_vendas,
                              right=df_produtos,
                              on='produto',
                              how='left')
df_vendas_mesclado
```

Utilizando o método .map


```python
dict_produtos = df_produtos.set_index('produto').to_dict()['preco']
dict_produtos
```

```python
df_vendas['preco_dict'] = df_vendas['produto'].map(dict_produtos)
df_vendas
```

Utilizando o método .replace

```python
del dict_produtos['Tenis Nike']
dict_produtos
```

```python
df_vendas['produto'].map(dict_produtos)
```

diferença para o map: o replace retorna o nome do produto (nesse caso) se ele não encontrar a correpondência, enquanto o map retorna o NaN, conforme se vê acima
