Estilizando a visualização

Formatando floats

```python
import pandas as pd
import numpy as np

df_tempo = pd.DataFrame(np.random.rand(10,2)*5,
                        index=pd.date_range(start='2023-01-01', periods=10),
                        columns=['Porto Alegre', 'São Paulo'])

df_tempo.style.format(decimal=',', precision=3)
```

Estilizando colunas

```python
df_tempo.style.format_index(str.upper, axis=1)
```

Estilizando o index

```python
df_tempo.style.format_index(lambda v: v.strftime('%d/%m/%Y'), axis=0)
```

Adicionando título

```python
df_tempo.style.set_caption('CONDIÇÕES DO TEMPO')
```

Adicionando gradiente (formatação condicional)

```python
import matplotlib
df_tempo.style.background_gradient(axis=None, vmin=1, vmax=5, cmap='YlGnBu')
```

Modificando valores

```python
def condicao_chuva(v):
    if v > 2.75:
        return 'Chuva Forte'
    elif v < 1.75:
        return "Seco"
    else: 
        return 'Chuva'
    
df_tempo.style.format(condicao_chuva)
```

Juntando tudo

```python
def condicao_chuva(v):
    if v > 2.75:
        return 'Chuva Forte'
    elif v < 1.75:
        return "Seco"
    else: 
        return 'Chuva'


def embelezar(styler):
    styler.set_caption('Condição do Tempo')
    styler.format(condicao_chuva)
    styler.format_index(lambda v: v.strftime('%d/%m/%Y'), axis=0)
    styler.background_gradient(axis=None, vmin=1, vmax=5, cmap='YlGnBu')
    return styler

df_tempo.style.pipe(embelezar)
```
