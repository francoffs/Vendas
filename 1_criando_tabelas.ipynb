### Criando a partir de listas

```python
import pandas as pd

data = [[1,2],
        [3,4],
        [5,6]
        ]
df = pd.DataFrame(data)
df
```

Criando a partir de dicionários de listas

```python
import pandas as pd

data = {'Nome':["Farnei", 'Martina', 'Gabriel'],
        'Sobrenome': ['Franco', 'Mohr', 'Siqueira']}
df = pd.DataFrame(data)
df
```

Criando com índice

```python
import pandas as pd

data = {'Nome':["Farnei", 'Martina', 'Gabriel'],
        'Sobrenome': ['Franco', 'Mohr', 'Siqueira']}
df = pd.DataFrame(data, index=['nome1', 'nome2', 'nome3'])
df
```

Criando a partir de uma lista de dicionários

```python
import pandas as pd

data = [{'Nome':"Farnei", 'Sobrenome': 'Franco'},
        {'Nome':"Martina", 'Sobrenome': 'Mohr'},
        {'Nome':"Gabriel", 'Sobrenome': 'Siqueira'}
        ]
df = pd.DataFrame(data)
df
```

Lendo arquivos


```python
import pandas as pd
import openpyxl

tabelas_vendas = pd.read_excel('vendas.xlsx')
tabelas_vendas
```

```python
import pandas as pd
import openpyxl

tabelas_vendas = pd.read_excel('vendas.xlsx', sheet_name='Sheet1')
tabelas_vendas
```

```python
import pandas as pd
import openpyxl

tabelas_vendas = pd.read_excel('vendas.xlsx', index_col=1)
tabelas_vendas
```

```python
import pandas as pd
import openpyxl

tabelas_vendas = pd.read_excel('vendas.xlsx', decimal=',', usecols=[0,1,2])
tabelas_vendas
```
